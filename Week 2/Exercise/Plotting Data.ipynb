```octave
t = [0:0.01:0.98];
y1 = sin(2*pi*4*t);
```

```octave
plot(t,y1)
```

```octave

y2 = cos(2*pi*4*t);
plot(t,y2);
```

```octave
plot(t,y1);
hold on; % hold on will keep the image, so we can plot the 2nd image on top of the 1st one.
plot(t,y2,'r');
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('my plot')
print -dpng 'myPlot.png'
```

```octave
close % close the plot
```

```octave
figure(1); plot(t,y1);
figure(2); plot(t,y2);
```
